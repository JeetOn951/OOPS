# **Theoretical Questions :**



### 1. **What is Object-Oriented Programming (OOP) ?**

- OOP is a way of writing and organizing code using the idea of "objects"—just like real-world things. Each object can have data (attributes) and actions (methods). OOP makes code more reusable, organized, and easier to manage.



### 2. **What is a class in OOP ?**

- Think of a **class** as a blueprint. Like a blueprint for building a car—it defines what a car should have (wheels, engine) and do (drive, brake). In code, a class defines how its objects will behave.



### 3. **What is an object in OOP ?**

- An **object** is the real version made from the blueprint. If `Car` is the class, then your Honda or BMW is an object. It has all the features defined in the class.



### 4. **What is the difference between abstraction and encapsulation ?**

* **Abstraction** is **hiding the details** and showing only the essentials (like using a smartphone without knowing how its circuits work).
* **Encapsulation** is **keeping the data safe** by wrapping it inside a class and controlling access (like putting valuables in a locker).



### 5. **What are dunder methods in Python?**

- "Dunder" stands for **double underscore**. Methods like `__init__`, `__str__`, and `__len__` are dunder methods. They add special behavior to objects and allow things like printing, comparisons, etc., to work naturally.


### 6. **Explain the concept of inheritance in OOP**

- **Inheritance** allows a class to get features from another class. For example, if `Bird` is a base class, `Parrot` can inherit from it and automatically get the ability to "fly" or "chirp".



### 7. **What is polymorphism in OOP?**

- **Polymorphism** means “many forms.” It allows different objects to respond to the same function in their own way. For example, both `Dog` and `Cat` might have a `speak()` method, but one says "Woof" and the other "Meow".



### 8. **How is encapsulation achieved in Python?**

- In Python:

    **i.** We use **underscore `_` or double underscore `__`** to mark variables as private.

    **ii.** We create **getters and setters** using methods or decorators like `@property` to safely access or update values.



### 9. **What is a constructor in Python?**


- The `__init__()` method is called **automatically when an object is created**. It’s used to **initialize attributes** of the object (like setting the name or age).



### 10. **What are class and static methods in Python?**

- * **@classmethod** receives the class (`cls`) as the first argument. It can **modify class-level data**.
  * **@staticmethod** doesn’t take `self` or `cls`. It behaves like a normal function inside the class, just grouped there for convenience.



### 11. **What is method overloading in Python?**
- Python doesn't support traditional overloading like C++ or Java. But we can fake it by using **default arguments or `*args`/`**kwargs`** to handle different scenarios.



### 12. **What is method overriding in OOP?**

- **Overriding** happens when a child class **redefines** a method of the parent class. It’s like a child saying, “I’ll do this my own way.”



### 13. **What is a property decorator in Python?**

- The `@property` decorator makes a method **behave like an attribute**. It’s used to access private variables in a clean way—without calling a function.



### 14. **Why is polymorphism important in OOP?**

- Polymorphism allows flexible and reusable code. You can write functions that work with **different object types**, as long as they follow the expected behavior (like all animals having a `speak()` method).



### 15. **What is an abstract class in Python?**

- An abstract class **can't be instantiated** (you can’t create an object of it). It’s like a template where **some methods must be implemented** by any subclass. You define it using `abc` module.



### 16. **What are the advantages of OOP?**

* Clean, modular structure
* Code reuse via inheritance
* Easier to maintain and debug
* Encapsulation for security
* Polymorphism for flexibility



### 17. **What is multiple inheritance in Python?**

- Multiple inheritance means a class can **inherit from more than one parent class**. Python supports this, though it can get tricky—Python resolves conflicts using the **Method Resolution Order (MRO)**.



### 18. **What is the difference between a class variable and an instance variable?**

* **Class variable**: Shared across all objects. (e.g., company name)
* **Instance variable**: Unique to each object. (e.g., employee name)



### 19. **Explain the purpose of `__str__` and `__repr__` methods in Python**

* `__str__`: Defines the **user-friendly** string (used by `print()`).
* `__repr__`: Defines a **developer-friendly** string (used in debugging or console).

Example:

```python
def __str__(self):
    return "User-friendly info"
def __repr__(self):
    return "Technical info"
```



### 20. **What is the significance of the `super()` function in Python?**

- `super()` allows you to **call methods of the parent class**, usually inside constructors or overridden methods. It helps in **code reuse and extending functionality**.



### 21. **What is the significance of the `__del__` method in Python?**

- `__del__()` is called when an object is **about to be destroyed**. It's like a destructor and is used to clean up resources, though it’s not commonly used in Python.



### 22. **What is the difference between @staticmethod and @classmethod in Python?**

| Feature        | `@staticmethod` | `@classmethod`                       |
| -------------- | --------------- | ------------------------------------ |
| First argument | None            | Class (`cls`)                        |
| Access class?  | No              | Yes                                  |
| Use case       | Utility methods | Factory methods or alter class state |



### 23. **How does polymorphism work in Python with inheritance?**

- When multiple child classes inherit from the same parent and override its methods, you can **call the same method on different objects**, and each will respond differently. Python uses **dynamic typing** to handle this.

Example:

```python
for animal in [Dog(), Cat()]:
    animal.speak()  # Different results, same method call
```



### 24. **What is method chaining in Python OOP?**

- Method chaining means calling **multiple methods on the same object in a single line**. To enable this, methods must return `self`.

Example:

```python
car.start().accelerate().brake()
```



### 25. **What is the purpose of the `__call__` method in Python?**

- If a class has a `__call__()` method, then its **objects behave like functions**.

Example:

```python
class Greet:
    def __call__(self, name):
        print(f"Hello, {name}")

g = Greet()
g("Alice")  # Acts like a function
```

````python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
circle = Circle(5)
print(f"Area of Circle: {circle.area()}")

rectangle = Rectangle(4, 6)
print(f"Area of Rectangle: {rectangle.area()}")
````

````python

# 3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
#and further derive a class Electric Car that adds a battery attribute.


class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, model):
        super().__init__(vehicle_type)
        self.model = model

class ElectricCar(Car):
    def __init__(self, vehicle_type, model, battery_capacity):
        super().__init__(vehicle_type, model)
        self.battery_capacity = battery_capacity

    def display_info(self):
        print(f"Vehicle Type: {self.type}, Model: {self.model}, Battery Capacity: {self.battery_capacity} kWh")

# Example usage
my_electric_car = ElectricCar("Electric Car", "Tesla Model 3", 75)
my_electric_car.display_info()
````

````python
# 4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("Birds can generally fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrows can fly short distances")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly, but they can swim")

# Demonstrated polymorphism
def bird_flight_test(bird):
    bird.fly()

print("Testing polymorphism:")
sparrow = Sparrow()
penguin = Penguin()

bird_flight_test(sparrow)
bird_flight_test(penguin)
````

````python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute using name mangling

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if self.__balance >= amount:
                self.__balance -= amount
                print(f"Withdrew: ${amount}. New balance: ${self.__balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def check_balance(self):
        print(f"Current balance: ${self.__balance}")

# Example usage
account = BankAccount(1000)
account.check_balance()
account.deposit(500)
account.withdraw(200)
account.check_balance()
account.withdraw(1500) # Attempt to withdraw more than balance
````

````python
#6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
class Instrument:
    def play(self):
        print("Playing an instrument")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar")

class Piano(Instrument):
    def play(self):
        print("Playing the piano keys")

# Demonstrate runtime polymorphism
def make_instrument_play(instrument):
    instrument.play()

print("Demonstrating runtime polymorphism:")
guitar = Guitar()
piano = Piano()
instrument = Instrument()

make_instrument_play(guitar)
make_instrument_play(piano)
make_instrument_play(instrument)
````

````python
# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, x, y):
        return x + y

    @staticmethod
    def subtract_numbers(x, y):
        return x - y

# Example usage
result_add = MathOperations.add_numbers(10, 5)
print(f"Addition result: {result_add}")

result_subtract = MathOperations.subtract_numbers(10, 5)
print(f"Subtraction result: {result_subtract}")
````

````python
# 8. Implement a class Person with a class method to count the total number of persons created.

class Person:
    total_persons = 0  # Class variable to count instances

    def __init__(self, name):
        self.name = name
        Person.total_persons += 1 # Increment the class variable

    @classmethod
    def get_total_persons(cls):
        return cls.total_persons

# Example usage
person1 = Person(" Abhoy")
person2 = Person(" Soumadeep")
person3 = Person("Sounak")

print(f"Total number of persons created: {Person.get_total_persons()}")
````

````python
# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        else:
            raise TypeError("Can only add a Vector to another Vector")

# Example usage
v1 = Vector(2, 3)
v2 = Vector(1, 4)

v3 = v1 + v2
print(v3)

# Example of adding with a non-Vector (will raise TypeError)
# v4 = v1 + 5
# print(v4)
````

````python
#  12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # Assuming grades is a list of numbers

    def average_grade(self):
        if not self.grades:
            return 0  # Return 0 if there are no grades
        return sum(self.grades) / len(self.grades)

# Example usage
student1 = Student("Bob", [85, 90, 78, 92])
print(f"{student1.name}'s average grade: {student1.average_grade()}")

student2 = Student("Alice", [])
print(f"{student2.name}'s average grade: {student2.average_grade()}")
````

````python
#   13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.

class Rectangle:
    def __init__(self):
        self.width = 0
        self.height = 0

    def set_dimensions(self, width, height):
        if width >= 0 and height >= 0:
            self.width = width
            self.height = height
        else:
            print("Dimensions must be non-negative.")

    def area(self):
        return self.width * self.height

# Example usage
rectangle = Rectangle()
rectangle.set_dimensions(5, 10)
print(f"Area of the rectangle: {rectangle.area()}")

rectangle.set_dimensions(-2, 5) # Example of invalid dimensions
print(f"Area of the rectangle: {rectangle.area()}") # Area remains unchanged
````

````python
#  14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked  and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate

    def calculate_salary(self, hours_worked):
        return self.hourly_rate * hours_worked

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self, hours_worked):
        base_salary = super().calculate_salary(hours_worked)
        return base_salary + self.bonus

# Example usage
employee = Employee(" Sayan", 20)
print(f"{employee.name}'s salary: {employee.calculate_salary(40)} rupees")

manager = Manager(" Sporsho", 25, 500)
print(f"{manager.name}'s salary:   {manager.calculate_salary(40)} rupees")
````

````python
#  16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo!"

class Sheep(Animal):
    def sound(self):
        return "Baa!"

# Example usage
cow = Cow()
print(f"Cow says: {cow.sound()}")

sheep = Sheep()
print(f"Sheep says: {sheep.sound()}")

# Attempting to instantiate the abstract class will raise an error
# generic_animal = Animal()
# generic_animal.sound()
````

````python
#  17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Example usage
book1 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 1979)
print(book1.get_book_info())

book2 = Book("Pride and Prejudice", "Jane Austen", 1813)
print(book2.get_book_info())
````

````python
# 18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Example usage
house = House("Rajnagar 3, Teliamura", 30000000)
print(f"House at {house.address} costs  {house.price} rupees")

mansion = Mansion("Shivnagar, Agartala ", 15000000, 10)
print(f"Mansion at {mansion.address} costs  {mansion.price} rupees and has {mansion.number_of_rooms} rooms.")
````
